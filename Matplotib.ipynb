##Data Visualization with Python using Matplotlib

```python
import matplotlib.pyplot as plt

# Sample data for the plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

# function we used creates a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My First Matplotlib Plot")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data for the scatter plot
np.random.seed(0)  # for reproducibility
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(20, 100, 50)

# Create the scatter plot
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, marker='o')

# Add labels and title
plt.xlabel("Random X-axis Values")
plt.ylabel("Random Y-axis Values")
plt.title("Scatter Plot with Random Data")

# Add a colorbar to show the color scale
plt.colorbar(label='Color Intensity')

# Add a grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for three different groups across four categories
categories = ['Category A', 'Category B', 'Category C', 'Category D']
group1_values = [25, 32, 34, 20]
group2_values = [10, 20, 25, 30]
group3_values = [15, 28, 30, 25]

# Set the width of the bars
bar_width = 0.25

# Set the positions of the bars on the x-axis
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the grouped bar chart
plt.bar(r1, group1_values, color='blue', width=bar_width, edgecolor='grey', label='Group 1')
plt.bar(r2, group2_values, color='red', width=bar_width, edgecolor='grey', label='Group 2')
plt.bar(r3, group3_values, color='green', width=bar_width, edgecolor='grey', label='Group 3')

# Add labels and title
plt.xlabel('Categories', fontweight='bold')
plt.ylabel('Values', fontweight='bold')
plt.title('Grouped Bar Chart')

# Set the x-axis ticks to be the categories
plt.xticks([r + bar_width for r in range(len(categories))], categories)

# Add a legend
plt.legend()

# Add a grid for better readability
plt.grid(axis='y', linestyle='--')

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the line plot
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)

# Sample data for the bar chart
categories = ['A', 'B', 'C', 'D']
values = [30, 45, 22, 58]

# Create a figure with two subplots (1 row, 2 columns)
fig = plt.figure(figsize=(10, 5))

# Add the first subplot (for the line plot)
ax1 = fig.add_subplot(1, 2, 1)  # (rows, columns, panel number)
ax1.plot(x1, y1, color='blue')
ax1.set_xlabel("Time")
ax1.set_ylabel("Sine Wave")
ax1.set_title("Line Plot")
ax1.grid(True)

# Add the second subplot (for the bar chart)
ax2 = fig.add_subplot(1, 2, 2)  # (rows, columns, panel number)
ax2.bar(categories, values, color='green')
ax2.set_xlabel("Categories")
ax2.set_ylabel("Values")
ax2.set_title("Bar Chart")

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the figure with subplots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
plt.figure(figsize=(12, 7))
plt.subplots_adjust(right=1)
plt.plot(x, y1, label='Sine Wave', color='blue', linestyle='-', marker='o', markersize=5)
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--', marker='^', markersize=5)

# Add labels and title with more styling
plt.xlabel("Time (seconds)", fontsize=14, fontweight='bold')
plt.ylabel("Amplitude", fontsize=14, fontweight='bold')
plt.title("Sine and Cosine Waves", fontsize=16, fontweight='bold')

# Add a more detailed legend with a title
plt.legend(title='Wave Types', fontsize=12)

# Add a grid with custom styling
plt.grid(True, linestyle=':', linewidth=0.5, color='gray')

# Annotate a specific point on the sine wave
annotation_x = 2 * np.pi  # Approximately one cycle
annotation_y = np.sin(annotation_x)
plt.annotate(f'Local Maxima at x={annotation_x:.2f}',
             xy=(annotation_x, annotation_y),
             xytext=(annotation_x + 2, annotation_y + 1),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10)

# Add text to the plot
plt.text(8, 0.5, 'This plot shows two fundamental trigonometric functions.', fontsize=11, style='italic', bbox=dict(facecolor='yellow', alpha=0.2))

# Improve plot aesthetics
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create the figure and add a 3D subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the data for the surface plot
num1 = np.arange(-5, 5, 0.25)
num2 = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(num1, num2)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add a title to the plot
ax.set_title('3D Surface Plot of a Sine Wave')

# Add a colorbar to the plot
fig.colorbar(surface, shrink=0.5, aspect=10)

# Improve the viewing angle (optional)
ax.view_init(30, 45) # Elevation and Azimuth

# Show the plot
plt.show()
```
